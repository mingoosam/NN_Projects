```python
import torch
import numpy as np

# create a list of tensors
tensors_list = [torch.tensor([1, 2, 3]), torch.tensor([4, 5, 6]), torch.tensor([7, 8, 9])]
tensors_list2 = [torch.tensor([1, 0, 3]), torch.tensor([4, 5, 0]), torch.tensor([0, 8, 9])]

# stack the tensors along a new dimension
stacked_tensor = torch.stack(tensors_list)
stacked_tensor2 = torch.stack(tensors_list2)


# convert the tensor to a numpy array
numpy_array = stacked_tensor.numpy()
numpy_array2 = stacked_tensor2.numpy()

print(numpy_array)
print(numpy_array2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import pickle

# Generate some example data
y_true = np.array([0, 1, 1, 0, 1, 0, 0, 1])
y_pred = np.array([1, 1, 0, 0, 1, 1, 0, 1])

# Create the confusion matrix
cm = confusion_matrix(y_true, y_pred)

# Create a display object
disp = ConfusionMatrixDisplay.from_predictions(y_true, y_pred, display_labels=['Class 0', 'Class 1'])

# Plot the confusion matrix
disp.plot()

plt.show()

y_true.shape
```
